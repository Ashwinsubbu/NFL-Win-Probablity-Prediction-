## Visualization of team performance

```python
teams=data['posteam'].unique()
teams
```

# Average Win Probability of teams 

```python
avg_winprob={}
sliced=data[data['posteam']==teams[0]]
avg_winprob[teams[0]]=sliced['Win_Prob'].mean()
avg_winprob
```

```python
avg_winprob={}
for team in teams:
    sliced=data[data['posteam']==team]
    avg_winprob[team]=sliced['Win_Prob'].mean()
```

```python
daf=pd.DataFrame([[key,value] for key,value in avg_winprob.items()],columns=["Team","Win_Prob"])
```

```python
df=daf[daf['Team'].isin(['PIT', 'TEN','CLE', 'MIN', 'NO', 'DET', 'DAL', 'TB', 'HOU',
       'NYJ', 'IND', 'JAC', 'DEN', 'CIN', 'CAR', 'PHI', 'KC', 'BAL', 'ATL',
       'MIA', 'ARI', 'SF', 'STL', 'SEA', 'NYG', 'WAS', 'GB', 'CHI', 'NE',
       'BUF', 'OAK', 'SD'])]
```

# Plot of Win probabilities 

```python
fig = df.plot.bar(x='Team',y='Win_Prob',color='blue',figsize=(20,10),fontsize=16)
fig.set_title("Effective Play Rankings", fontsize=20)
fig.set_xlabel('Team',fontsize=18)
fig.set_ylabel('Win Probability',fontsize=18)
```
